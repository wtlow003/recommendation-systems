# Split Train-Test Dataset

```python
# setting global variable
CATEGORY = "Pet_Supplies"

# split ratio
split_ratio = 0.7
```

```python
# load dataset
products = pd.read_csv(f"../../data/processed/{CATEGORY}_processed.csv")

# check first 5 rows
products.head()
```

```python
# convert reviewTime into datetime format
products["reviewTime"] = products["reviewTime"].progress_apply(lambda x: '/'.join(x.replace(',', '').split(' ')))
products["reviewTime"] = products["reviewTime"].progress_apply(lambda x: datetime.strptime(x, "%m/%d/%Y"))
```

```python
# we need to first group by reviewers and sort by each review's time each individual "groups"

reviews_by_users = (products.groupby(['reviewerID'])
                    .progress_apply(lambda x: x.sort_values(["reviewTime"], ascending=True))
                    .reset_index(drop=True))

reviews_by_users
```

```python
user_review_map = reviews_by_users.groupby(['reviewerID'])["index"].progress_apply(list).to_dict()
```

```python
user_review_map
```

```python
test_review_index = set()
# test_review_index = list()
for user in user_review_map:
    review_pos_cutoff = math.floor(len(user_review_map[user]) * split_ratio)
    test_review_index.update(user_review_map[user][review_pos_cutoff:])
    # test_review_index.extend(user_review_map[user][review_pos_cutoff:])
```

```python
# generate train test
train = products[~products["index"].isin(test_review_index)]
test = products[products["index"].isin(test_review_index)]
```

```python
# removing any users/items that is in test but NOT in train
train_prods = train["asin"].unique()
train_users = train["reviewerID"].unique()

test = test[
    (test["asin"].isin(train_prods)) & (test["reviewerID"].isin(train_users))
]
```

```python
# summary statistics
print(f"For trainset, uniquely: {train['reviewerID'].nunique()} users, {train['asin'].nunique()} products, {train.shape} RxC.")
print(f"For testset, uniquely: {test['reviewerID'].nunique()} users, {test['asin'].nunique()} products, {test.shape} RxC.")
```

```python
# drop existing `index` column and reset index
train, test = train.drop(["index"], axis=1), test.drop(["index"], axis=1)
```
